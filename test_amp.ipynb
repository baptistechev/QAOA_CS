```python
import numpy as np
import matplotlib.pyplot as plt

l = []
for x,y in zip(np.random.uniform(0,0.5,20),np.random.uniform(0,0.5,20)):
    l.append(complex(x,y))
l = np.array(l)/ np.sqrt(np.sum([(z.real**2+z.imag**2) for z in l]))
l = sorted(l, key=lambda z: (z.real**2+z.imag**2))

real_part = [z.real for z in l]
imag_part = [z.imag for z in l]

plt.bar(np.arange(len(l)),[(z.real**2+z.imag**2) for z in l])
plt.show()

plt.bar(np.arange(len(l)),real_part)
plt.show()

plt.bar(np.arange(len(l)),imag_part)
plt.show()
```

```python
import copy

l_prime = copy.deepcopy(l)

for _ in range(50):
    i = np.random.randint(len(l_prime))
    zi = l_prime[i]
    print(i,zi)

    l_prime = [z for z in l_prime if (z.real>=zi.real or z.imag>=zi.imag)]

    if len(l_prime)==1:
        print(l_prime[0])
        break

real_part_prime = [z.real for z in l_prime]
imag_part_prime = [z.imag for z in l_prime]

plt.bar(np.arange(len(l_prime)),[(z.real**2+z.imag**2) for z in l_prime])
plt.show()

plt.bar(np.arange(len(l_prime)),real_part_prime)
plt.show()

plt.bar(np.arange(len(l_prime)),imag_part_prime)
plt.show()
```
